```python
#code for gradient descent
import numpy as np
def gd(x,y):
  m_curr=b_curr=0
  n=len(x)
  lr=0.01

  for i in range(1000):
    y_pred=m_curr*x+b_curr
    loss=(1/n)*sum([val**2 for val in (y-y_pred)])
    md=-(2/n)*sum(x*(y-y_pred))
    bd=-(2/n)*sum(y-y_pred)
    m_curr=m_curr-lr*md
    b_curr=b_curr-lr*bd
    print("m {},b {},loss {} in iteration {}".format(m_curr,b_curr,loss,i))
```

```python
x=np.array([1,2,3,4,5])
y=np.array([6,7,8,9,10])
gd(x,y)
```
