# Encoding

```python
from sklearn.preprocessing import LabelEncoder
# Encoding the category column
le = LabelEncoder()
df["Category"] =  le.fit_transform(df["Category"])
df["Leather interior"] = le.fit_transform(df["Leather interior"])
df["Fuel type"] = le.fit_transform(df["Fuel type"])
df["Gear box type"] = le.fit_transform(df["Gear box type"])
df["Drive wheels"] = le.fit_transform(df["Drive wheels"])
```

```python
goofy = df["Mileage"]
```

```python
def remove_km(goofy):
  goofy = goofy.replace("km","")
  return goofy

df["Mileage"] = df["Mileage"].apply(remove_km)
df["Mileage"] = df["Mileage"].astype(int)
```

```python
def categorize_price(x):
  if x > 0 and x <10000:
    return 0
  elif x > 10000 and x < 20000:
    return 1
  elif x > 20000 and x < 30000:
    return 2
  else :
    return 3
```

```python
df["new Price"]=df["Price"].apply(categorize_price)
```
